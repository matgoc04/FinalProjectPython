```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('cleanCorgi.csv')
df = df.drop(columns='Unnamed: 0')
df.info()
```

```python
df_comments = df[df['comments_full'] != '[]']
df_comments[['text', 'comments_full', 'comments']]
df_comments['comments_full'].iloc[0]
```

```python
df_comments
```

```python
import datetime
import json
comments = []
for i in range(df_comments.shape[0]):
    x = eval(df_comments['comments_full'].iloc[i])
    for cmt in x:
        comments.append(cmt['comment_text'])
comments
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Gá»™p táº¥t cáº£ cÃ¡c comments thÃ nh má»™t chuá»—i vÄƒn báº£n
all_comments_text = ' '.join(comments)

# Táº¡o WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_comments_text)

# Hiá»ƒn thá»‹ WordCloud
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud cá»§a Comments')
plt.show()
# tháº¥y tá»« xuáº¥t hiá»‡n nhiá»u nháº¥t trong cÃ¡c comments cÃ³ váº» lÃ  Nguyá»…n, Ä‘Ã¢y lÃ  há» tÃªn ngÆ°á»i, váº­y cÃ³ kháº£ nÄƒng cao cÃ¡c comments sáº½ nhiá»u á»Ÿ viá»‡c tag ngÆ°á»i thÃ¢n báº¡n bÃ¨
```
