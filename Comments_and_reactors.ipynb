```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('cleanCorgi.csv')
df = df.drop(columns='Unnamed: 0')
df.info()
```

```python
df_comments = df[df['comments_full'] != '[]']
df_comments[['text', 'comments_full', 'comments']]
df_comments['comments_full'].iloc[0]
```

```python
df_comments
```

```python
import datetime
import json
comments = []
for i in range(df_comments.shape[0]):
    x = eval(df_comments['comments_full'].iloc[i])
    for cmt in x:
        comments.append(cmt['comment_text'])
comments
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# G·ªôp t·∫•t c·∫£ c√°c comments th√†nh m·ªôt chu·ªói vƒÉn b·∫£n
all_comments_text = ' '.join(comments)

# T·∫°o WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_comments_text)

# Hi·ªÉn th·ªã WordCloud
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud c·ªßa Comments')
plt.show()
# th·∫•y t·ª´ xu·∫•t hi·ªán nhi·ªÅu nh·∫•t trong c√°c comments c√≥ v·∫ª l√† Nguy·ªÖn, ƒë√¢y l√† h·ªç t√™n ng∆∞·ªùi, v·∫≠y c√≥ kh·∫£ nƒÉng cao c√°c comments s·∫Ω nhi·ªÅu ·ªü vi·ªác tag ng∆∞·ªùi th√¢n b·∫°n b√®
```

```python
import pandas as pd

df_reactors = df[df['reactors'] != '[]']
list_reactors = df_reactors['reactors'].apply(eval).tolist()
df_reactors_full = pd.DataFrame()
for i in list_reactors:
   df_reactors_full = pd.concat([df_reactors_full, pd.DataFrame(i)], ignore_index=True)
df_reactors_full
```
